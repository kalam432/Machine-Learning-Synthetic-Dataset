```python
import pandas as pd
import random


input_file = "std data.xlsx"
input_data = pd.read_excel(input_file)

names = input_data["NAME"].tolist()  
emails = input_data["EMAIL ID"].tolist()  


course_names = ["Math", "Physics", "Chemistry", "Biology", "Computer Science", "History"]
attempts = ["First", "Second", "Third"]

used_course_ids = set()  
def assign_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    elif marks >= 60:
        return "D"
    elif marks >= 50:
        return "E"
    else:
        return "F"


def generate_course_id():
    while True:
        course_id = f"C-{random.randint(1000, 9999)}"
        if course_id not in used_course_ids:
            used_course_ids.add(course_id)
            return course_id


students = []
for i in range(len(names)):  
    num_courses = random.randint(1, 3)
 
    age = random.randint(18, 25)  
    for _ in range(num_courses):
        course_name = random.choice(course_names)
        attempt = random.choice(attempts)

        if attempt == "First":
            marks = random.randint(40, 100)
        elif attempt == "Second":
            marks = random.randint(50, 100)
        else:
            marks = random.randint(60, 100)

        grade = assign_grade(marks)

        student = {
            "Name": names[i],
            "Email": emails[i],
            "Age": age,
            "Course ID": generate_course_id(),
            "Course Name": course_name,
            "Attempt": attempt,
            "Marks": marks,
            "Grade": grade
        }
        students.append(student)


df = pd.DataFrame(students)

# Save to CSV
output_file = "enhanced_student_dataset.csv"
df.to_csv(output_file, index=False)

print(f"Dataset created and saved as '{output_file}'.")


print(df.head())
df
```
